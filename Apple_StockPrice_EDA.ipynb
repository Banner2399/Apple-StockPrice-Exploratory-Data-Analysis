Data Pre-Processing

```python
#Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('AppleStockPrice.csv', encoding='unicode_escape')

df.head()
```

```python
df.shape

df.info()
```

Data Cleaning

```python
#checking for null values

df.isnull().sum()
```

Data Visualization

```python
plt.figure(figsize=(6,6))
df.dtypes.value_counts().plot.pie(autopct = '%0.2f%%')

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.stem(list(df['Close'].value_counts().keys()))
plt.xlabel("Quantity Times of Close", fontsize=15)
plt.ylabel("Price of Close Share", fontsize=15)
plt.title("Close Price of Share's Graph", fontsize=15)
plt.show()

df['Close'].value_counts().sum()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.plot(list(df['High'].value_counts().keys()))
plt.title("High's plot")
plt.subplot(132)
plt.plot(list(df['Low']))
plt.title("Low's Plot")
plt.subplot(133)
plt.plot(list(df['Adj Close']))
plt.title("Adj Close's Plot")
plt.suptitle("Graphical Representation of High and Low Share Price and Adj Close")

plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.hist(list(df['High'].value_counts().keys()))
plt.title("Histogram")
plt.subplot(132)
plt.hist(list(df['Low']))
plt.title("Historgram")
plt.subplot(133)
plt.hist(list(df['Adj Close']))
plt.title("Histogram")
plt.suptitle("Graphical Representation of High and Low Share Price and Adj Close")

plt.show()
```

```python
df.plot(subplots=True, figsize=(15,12))
plt.show()
```
